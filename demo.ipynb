# Import the Custom Library & Create Generator Object

Import the custom library called generateDistribution from the local generateDistribution.py

```python
import generateDistribution as genDist
```

Create a distribution Generator Object

```python
dist = genDist.distGenerator()
```

# Generate Uniform Random Variates

Use scipy uniforms to make our custom generated distributions comparable to scipy generated distributions in order to make sure they are transformation of the same set of uniform variables

```python
from scipy import stats
import numpy as np
size = 1000
seed = 123

unif = stats.uniform.rvs(size=size,random_state=seed)
dist.plotDist('Uniform',unif)
```

# Generate Exponential Random Variates

Generate Random Variables using the Library and Plot the Generated Variables

```python
# Let's use Lambda = 2
Lambda = 2
Dist_Name = f'Exponential(Lambda={Lambda})'

# Generate Random Variables using the Library
exp = dist.generateExponential(Lambda=Lambda, size=size, random_state=seed)

# Plot the Generated Distribution
dist.plotDist(Dist_Name, exp)
```

Validate our generated distribution is the same as that generated from Scipy

```python
# Reference Distribution is from Scipy
exp_ref = stats.expon.rvs(scale=(1/Lambda), size=size, random_state=seed)

# Comparative Plot of Generated vs Scipy Reference Distribution
dist.compPlotDist(Dist_Name,exp,exp_ref)

# Validate Generated Distribution vs the Scipy Distribution
dist.validateDist(Dist_Name,exp, exp_ref,alpha=0.05)
```
